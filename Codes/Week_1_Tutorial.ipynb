## Part 3: Lists - Storing Multiple Values

### 3.1 Creating Lists

Lists store multiple values in order. They use square brackets `[]`.

````python
# A list of monthly returns (%)
returns = [2.5, -1.2, 3.8, 0.5, -2.1, 1.9]
print(returns)
````

````python
# How many items in the list?
print(f"Number of months: {len(returns)}")
````

````python
# Access individual items (counting starts at 0!)
print(f"First month return: {returns[0]}%")
print(f"Second month return: {returns[1]}%")
print(f"Last month return: {returns[-1]}%")  # -1 means last item
````

⚠️ **Important:** Python counts from 0, not 1. The first item is at position `[0]`, the second at `[1]`, etc.

### 3.2 Simple List Operations

````python
# Sum of all returns
total_return = sum(returns)
print(f"Total return over 6 months: {total_return}%")
````

````python
# Average return
average_return = sum(returns) / len(returns)
print(f"Average monthly return: {average_return:.2f}%")
````

````python
# Maximum and minimum
print(f"Best month: {max(returns)}%")
print(f"Worst month: {min(returns)}%")
````

💡 **Tip:** The `:.2f` in the print statement means "format as a number with 2 decimal places".

## Part 4: Using Libraries

### 4.1 What Are Libraries?

Libraries are collections of pre-written code that extend Python's capabilities. For data analysis, we use:

- **NumPy**: Fast numerical computations
- **Pandas**: Data manipulation (like Excel, but more powerful)
- **Matplotlib**: Creating charts and plots

We load libraries using `import`.

````python
# Import libraries
import numpy as np      # 'np' is a shorthand nickname
import pandas as pd     # 'pd' is standard for pandas
import matplotlib.pyplot as plt  # 'plt' for plotting

print("Libraries loaded successfully!")
````

### 4.2 NumPy Arrays

NumPy arrays are like lists, but much faster and with more built-in functions.

````python
# Convert our list to a NumPy array
returns_array = np.array(returns)
print(returns_array)
````

````python
# NumPy has built-in statistical functions
print(f"Mean: {np.mean(returns_array):.2f}%")
print(f"Standard deviation: {np.std(returns_array):.2f}%")
print(f"Median: {np.median(returns_array):.2f}%")
````

## Part 6: Creating Visualizations

### 6.1 Line Plot

````python
# Plot market returns over time
plt.figure(figsize=(12, 4))  # Set the size (width, height)
plt.plot(df['MktRf'])
plt.title('Monthly Market Excess Returns')
plt.xlabel('Observation Number')
plt.ylabel('Return (%)')
plt.axhline(y=0, color='red', linestyle='--', alpha=0.5)  # Add a horizontal line at 0
plt.show()
````

### 6.2 Histogram

A histogram shows the distribution of values.

````python
plt.figure(figsize=(10, 4))
plt.hist(df['MktRf'], bins=50, edgecolor='white')
plt.title('Distribution of Monthly Returns')
plt.xlabel('Return (%)')
plt.ylabel('Frequency')
plt.axvline(x=df['MktRf'].mean(), color='red', linestyle='--', label='Mean')
plt.legend()
plt.show()
````

### 6.3 Scatter Plot

A scatter plot shows the relationship between two variables.

````python
plt.figure(figsize=(8, 6))
plt.scatter(df['d_p'], df['MktRf'], alpha=0.3, s=10)
plt.title('Dividend-Price Ratio vs Market Return')
plt.xlabel('Dividend-Price Ratio (d_p)')
plt.ylabel('Market Return (%)')
plt.show()
````

## Part 8: Connecting to Lecture Concepts

### 8.1 Why Time Order Matters

In Lecture 1, we discussed that financial data is **ordered in time**. Let's see this in practice.

````python
# The DATE column shows our data is ordered chronologically
print("First observation:", df['DATE'].iloc[0])
print("Last observation:", df['DATE'].iloc[-1])
````

### 8.2 The Prediction Problem

Our goal in this course: Can we use today's information to predict **next month's** returns?

This is fundamentally different from:
- Explaining past returns (econometrics)
- Understanding relationships (statistics)

````python
# Simple demonstration: correlation between d_p and CURRENT returns
corr_same = df['d_p'].corr(df['MktRf'])
print(f"Correlation (d_p vs same-month return): {corr_same:.3f}")

# But what about NEXT month's returns?
df['MktRf_next'] = df['MktRf'].shift(-1)  # Shift returns up by 1 row
corr_next = df['d_p'].corr(df['MktRf_next'])
print(f"Correlation (d_p vs next-month return): {corr_next:.3f}")
````

💡 **Key Insight:** The correlation with *future* returns is what matters for prediction, and it's typically much smaller than correlations with current returns.
